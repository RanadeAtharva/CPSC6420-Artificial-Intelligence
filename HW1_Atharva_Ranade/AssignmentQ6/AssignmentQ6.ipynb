```python
import numpy as np
up_cost= 0
left_cost=0
right_cost=0
down_cost=0
class MyNode():
    def __init__(self,s,p_node,move,depth,step_cost,path_cost):
        self.s = s
        self.p_node = p_node
        self.move = move
        self.depth = depth
        self.step_cost = step_cost
        self.path_cost = path_cost
        self.move_up = None
        self.move_left = None
        self.move_right = None
        self.move_down = None
    
    def try_down(self):
        z_index = [i[0] for i in np.where(self.s==0)]
        if z_index[0] == 0:
            return False
        else:
            u_value = self.s[z_index[0]-1,z_index[1]]
            new_s = self.s.copy()
            new_s[z_index[0],z_index[1]] = u_value
            new_s[z_index[0]-1,z_index[1]]= 0
            return new_s,u_value
    
    def try_right(self):
        z_index = [i[0] for i in np.where(self.s==0)]
        if z_index[1] == 0:
            return False
        else:
            l_value = self.s[z_index[0],z_index[1]-1]
            new_s = self.s.copy()
            new_s[z_index[0],z_index[1]] = l_value
            new_s[z_index[0],z_index[1]-1]= 0
            return new_s,l_value
    
    def try_left(self):
        z_index = [i[0] for i in np.where(self.s==0)]
        if z_index[1] == 2:
            return False
        else:
            r_value = self.s[z_index[0],z_index[1]+1]
            new_s = self.s.copy()
            new_s[z_index[0],z_index[1]] = r_value
            new_s[z_index[0],z_index[1]+1]= 0
            return new_s,r_value
        
    def try_up(self):
        z_index = [i[0] for i in np.where(self.s==0)]
        if z_index[0] == 2:
            return False
        else:
            d_value = self.s[z_index[0]+1,z_index[1]]
            new_s = self.s.copy()
            new_s[z_index[0],z_index[1]] = d_value
            new_s[z_index[0] +1,z_index[1]]= 0
            return new_s,d_value

    def path(self):
        s_trace=[self.s]
        move_trace=[self.move]
        depth_trace=[self.depth]
        step_cost_trace = [self.step_cost]
        path_cost_trace = [self.path_cost]
        while self.p_node:
            self = self.p_node
            s_trace.append(self.s)
            move_trace.append(self.move)
            depth_trace.append(self.depth)
            step_cost_trace.append(self.step_cost)
            path_cost_trace.append(self.path_cost)
        step_counter=0
        while s_trace:
            print ('step',step_counter)
            print( s_trace.pop())
            print("action=" ,move_trace.pop(),",depth=", str(depth_trace.pop()),", step cost=" , str(step_cost_trace.pop()),", total cost=", str(path_cost_trace.pop()))
            step_counter +=1
    def BFS(self,goal_s):
        queue = [self] 
        queue_num_nodes_popped = 0
        queue_max_length = 1
        depth_queue = [0]
        path_cost_queue = [0]
        visited = set([])
        
        while queue:
            if len(queue) > queue_max_length:
                  queue_max_length = len(queue)
            current_node = queue.pop(0)
            queue_num_nodes_popped += 1
            current_depth = depth_queue.pop(0)
            current_path_cost = path_cost_queue.pop(0)
            visited.add(tuple(current_node.s.reshape(1,9)[0]))
                
            if np.array_equal(current_node.s,goal_s):
                current_node.path()
                return True
            else:
                if current_node.try_down():
                    new_s,u_value = current_node.try_down()
                    if tuple(new_s.reshape(1,9)[0]) not in visited:
                        current_node.move_down = MyNode(s=new_s,p_node=current_node,move='up',depth=current_depth+1,step_cost=u_value,path_cost=current_path_cost+u_value)
                        queue.append(current_node.move_down)
                        depth_queue.append(current_depth+1)
                        path_cost_queue.append(current_path_cost+u_value)

                if current_node.try_right():
                    new_s,l_value = current_node.try_right()
                    if tuple(new_s.reshape(1,9)[0]) not in visited:
                        current_node.move_right = MyNode(s=new_s,p_node=current_node,move='left',depth=current_depth+1,step_cost=l_value,path_cost=current_path_cost+l_value)
                        queue.append(current_node.move_right)
                        depth_queue.append(current_depth+1)
                        path_cost_queue.append(current_path_cost+l_value)

                if current_node.try_up():
                    new_s,d_value = current_node.try_up()
                    if tuple(new_s.reshape(1,9)[0]) not in visited:
                        current_node.move_up = MyNode(s=new_s,p_node=current_node,move='down',depth=current_depth+1,step_cost=d_value,path_cost=current_path_cost+d_value)
                        queue.append(current_node.move_up)
                        depth_queue.append(current_depth+1)
                        path_cost_queue.append(current_path_cost+d_value)

                if current_node.try_left():
                    new_s,r_value = current_node.try_left()
                    if tuple(new_s.reshape(1,9)[0]) not in visited:
                        current_node.move_left = MyNode(s=new_s,p_node=current_node,move='right',depth=current_depth+1,step_cost=r_value,path_cost=current_path_cost+r_value)
                        queue.append(current_node.move_left)
                        depth_queue.append(current_depth+1)
                        path_cost_queue.append(current_path_cost+r_value)
                    
   
```
