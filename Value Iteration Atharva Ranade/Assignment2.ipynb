```python
class State:    
    def __init__(self, state, restricted, value, reward, optimum_policy, optimum_s_prime):
        self.state = state  
        self.x = state[0]
        self.y = state[1]
        self.d = state[2]
        self.restricted = restricted
        self.value = value
        self.reward = reward
        self.optimum_policy = optimum_policy
        self.optimum_s_prime = optimum_s_prime
```

```python
def max_value(state, list_of_states, noise, gamma):           
    possible_actions = []
    possible_s_prime = []
    possible_s_prime_negative_reward = []
    for A in possible_directions:
        new_state, negative_reward = take_action(state, A, list_of_states)
        if negative_reward != 0:
            possible_s_prime.append(new_state)
            possible_s_prime_negative_reward.append(negative_reward)
            possible_actions.append(A)
    
    number_of_unexpected_actions = (len(possible_s_prime)) -1


    values_list = []
    for new_state, negative_reward in zip(possible_s_prime, possible_s_prime_negative_reward):
       
        new_value = ( (1-noise)*((-negative_reward) + (gamma * new_state.value)) ) 
        

        unexpected_list_of_states = []
        unexpected_negative_rewards_list = []
        for unexpected, unexpected_negative_reward in zip(possible_s_prime, possible_s_prime_negative_reward):
            if unexpected == new_state:
                continue
            else:
                unexpected_list_of_states.append(unexpected)
                unexpected_negative_rewards_list.append(unexpected_negative_reward)
        for unexpected, unexpected_negative_reward in zip(unexpected_list_of_states, unexpected_negative_rewards_list):
            new_value += (noise/number_of_unexpected_actions)*((-unexpected_negative_reward) + (gamma * unexpected.value))

        values_list.append(new_value)

    max_value = -100000000000
    optimum_policy = ""
    optimum_s_prime = ""
    for value, action, s_prime in zip(values_list, possible_actions, possible_s_prime):
        if value > max_value:
            max_value = value
            optimum_policy = action
            optimum_s_prime = s_prime
            
    return max_value, optimum_policy, optimum_s_prime
```

```python
def update_state_val(new_list_of_states, goal_state, goal_state_reward, gameover_state, gameover_state_penalty, noise, gamma):
    previous_state_list = []
    previous_state_list = new_list_of_states[:]
    new_list_of_states = []
    i =0
    for state in previous_state_list:
        if state.restricted == 1:
            new_list_of_states.append(copy.deepcopy(state))
            continue
        elif (state.x == goal_state[0]) and (state.y == goal_state[1]):
            state.value = goal_state_reward
            new_list_of_states.append(copy.deepcopy(state))
            continue
        elif  (state.x == gameover_state[0]) and (state.y == gameover_state[1]):
            state.value = gameover_state_penalty
            new_list_of_states.append(copy.deepcopy(state))
            continue
        
        new_list_of_states.append(copy.deepcopy(state))
        new_list_of_states[-1].value, new_list_of_states[-1].optimum_policy, new_list_of_states[-1].optimum_s_prime = max_value(state, previous_state_list, noise, gamma)

        i+=1

    return previous_state_list, new_list_of_states
```

```python
def val_diff(previous_list_of_states, list_of_states):

    max_diff = 0
    for previous, state in zip(previous_list_of_states, list_of_states):
        
        if (abs(previous.value -state.value)) > max_diff:
            max_diff = abs(state.value - previous.value)
    
    return max_diff
```

```python
possible_directions = ["A1", "A2", "A3", "A4"]        
blocked_states = [[2,2], [3,2], [2,3]]                   
goal_state = [5,5]                                       
goal_state_reward = 100                                  
gameover_state = [4,4]                                    
gameover_state_penalty = -1000
list_of_states = []
initial_value=0
obstacles_list = [["row", 5, 2, 3], ["col", 5, 3, 4]]       

#generating the states
for x in range(1,6):
    for y in range(1,6):
        for d in range(1,5):        
            block_flag = 0
            for restricted in blocked_states:
                if (x == restricted[0]) and (y == restricted[1]):
                    block_flag = 1
                if ((x == goal_state[0]) and (y == goal_state[1])):
                    gen_state = State([x,y,d], block_flag, 0, goal_state_reward, 0, 0)
                elif ((x == gameover_state[0]) and (y == gameover_state[1])):
                    gen_state = State([x,y,d], block_flag, 0, gameover_state_penalty, 0, 0)
                else: 
                    gen_state = State([x,y,d], block_flag, 0, initial_value, 0, 0)
                list_of_states.append(gen_state)


```
